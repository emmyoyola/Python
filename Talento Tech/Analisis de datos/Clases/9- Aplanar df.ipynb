# Aplanar Data Frame
Reorganizarlo para que toda la información quede en una estructura más simple y de una sola capa. 

```python
import pandas as pd

# Tu DataFrame original
data = {
    'id': ['a1', 'a2', 'a3', 'a4', 'a5', 'a6', 'a7'],
    'abecedario': ['a', 'b', 'c', 'd', 'e', 'f', 'g'],
    'número': [1, 2, 3, 4, 5, 6, 7],
    'día': ['lunes', 'martes', 'miércoles', 'jueves', 'viernes', 'sábado', 'domingo'],
    'mes': ['enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio']
}

df = pd.DataFrame(data)
df
```

```python
# Usar pd.melt para transformar el DataFrame
df_melted = pd.melt(df, id_vars=['id'], value_name='Valor', var_name='Encabezado')

df_melted
```
