```python
import pandas as pd
```

```python
data18 = pd.read_csv(r'C:\Users\EMMY\Desktop\Talento Tech\Proyecto\DB Original\Evaluaciones_Agropecuarias_Municipales_EVA_2006-2018.csv')
data23 = pd.read_csv(r'C:\Users\EMMY\Desktop\Talento Tech\Proyecto\DB Original\Evaluaciones_Agropecuarias_Municipales___EVA._2019_-_2023._Base_Agr_cola_20240930 copy.csv')
df18 = data18
df23 = data23
```

```python
df18
```

```python
df18 = df18[(df18['CÓD. DEP.'] == 13) | (df18['CÓD. DEP.'] == 23)]
df18
```

```python
df18.columns
```

```python
df18 = df18.drop(['CÓD. DEP.', 'CÓD. MUN.', 'DESAGREGACIÓN REGIONAL Y/O SISTEMA PRODUCTIVO', 'PERIODO','ESTADO FISICO PRODUCCION', 'NOMBRE CIENTIFICO'], axis=1)
df23 = df23.drop(['Código Dane departamento', 'Código Dane municipio', 'Desagregación cultivo', 'Periodo', 'Estado físico del cultivo', 'Código del cultivo','Nombre científico del cultivo'], axis= 1)
```

```python
df18.columns
```

```python
df23.columns
```

```python
df18 = df18.map(lambda x: x.upper() if isinstance(x, str) else x)
df23 = df23.map(lambda x: x.upper() if isinstance(x, str) else x)
```

```python
# Crear un diccionario de mapeo para renombrar columnas en df18 basado en las columnas de df23
column_mapping = {
    'DEPARTAMENTO': 'Departamento',
    'MUNICIPIO': 'Municipio',
    'GRUPO DE CULTIVO': 'Grupo cultivo',
    'SUBGRUPO DE CULTIVO': 'Subgrupo',
    'CULTIVO': 'Cultivo',
    'AÑO': 'Año',
    'Área Sembrada (ha)': 'Área sembrada',
    'Área Cosechada(ha)': 'Área cosechada',
    'Producción(t)': 'Producción',
    'Rendimiento(t/ha)': 'Rendimiento',
    'CICLO DE CULTIVO': 'Ciclo del cultivo'
}
```

```python
# Renombrar las columnas de df18
df18.rename(columns=column_mapping, inplace=True)
```

```python
df18.columns
```

```python
df23.columns
```

```python
df_combined = pd.concat([df18, df23], ignore_index=True)
```

```python
df_combined.info()
```

```python
df_combined['Departamento'] = df_combined['Departamento'].replace({
    'BOLIVAR': 'BOLÍVAR',
    'CORDOBA': 'CÓRDOBA'
})
```

```python
df_combined['Departamento'].unique()
```

```python
df_combined.sort_values(['Departamento', 'Municipio'],inplace=True)
```

```python
df_combined.to_csv('Base Agrícola 2006-2023.csv', index=False)
```
