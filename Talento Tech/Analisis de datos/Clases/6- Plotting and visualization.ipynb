# Plotting and visualization whit Pandas

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Gr√°ficos de linea
Sirven para series de tiempo

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2024', 
periods=1000))
ts = ts.cumsum()    #Sumas acumuladas
ts.plot()  # Grafico de l√≠nea simple con pandas
```

```python
ts.head(2)
```

## Gr√°ficos de barra
Sirven para variables categ√≥ricas (preferiblemente)
* plot(kind='bar'): Barras verticales
* plot(kind='barh'): Barras horizontales

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=['A', 'B', 'C', 'D'])  # Gr√°fico de barras con pandas
df
```

```python
df.plot(kind='bar')
```

## Gr√°ficos de Dispersi√≥n (Scatter Plots)
Sirven para identificar tendencias entre 2 variables

```python
df.plot.scatter(x='A', y='B')
```

## Gr√°ficos de Caja (Box Plots)
Sirven para realizar comparaciones

```python
df.plot(kind='box')  # Gr√°fico de caja para mostrar la distribuci√≥n de las columnas
plt.show()
```

## Gr√°ficos de Histograma
Sirven para mostrar la distribuci√≥n de frecuencias

```python
df['A'].plot(kind='hist', bins=8)  #Histograma para la columna A
```

```python
df.plot(kind='line')
plt.title('Gr√°fico de L√≠nea')
plt.xlabel('√çndice')
plt.ylabel('Valores')
plt.legend(loc='best')
plt.grid(True)
```

## Subplot
Crea m√∫ltiples gr√°ficos en un solo lienzo

```python
df.plot(subplots=True, layout=(2, 2), figsize=(10, 8))
```

## Guardar Gr√°ficos

```python
df.plot(kind='line')
plt.savefig('grafico_linea.png')
```
