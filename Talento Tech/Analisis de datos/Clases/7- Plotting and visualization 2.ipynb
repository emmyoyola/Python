# Plotting and visualization whit Pandas 2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Semilla 
np.random.seed(42)

# Generar una base de datos aleatoria
data_size = 100
data = {
    'X': np.arange(data_size),
    'Y1': np.random.randint(1, 100, size=data_size),
    'Y2': np.random.normal(50, 15, size=data_size),
    'Category': np.random.choice(['A', 'B', 'C', 'D'],
size=data_size)
}

df = pd.DataFrame(data)
```

## Gráfico de linea

```python
#relación entre X y Y1
df.plot.line(y='Y1', x= 'X')
```

## Gráfico scatter

```python
# ¿Qué relación se visualiza en un gráfico de dispersión entre Y1 y Y2, y por qué es ideal para ver correlaciones?
df.plot.scatter(x='Y1', y='Y2')
```

## Gráfico de barras
*Recomendación:* hacer primero hacer la tabla de frecuencias

```python
#¿Cómo se compara el promedio de Y1 entre diferentes categorías en un gráfico de barras?
df.groupby('Category')['Y1'].mean().plot.bar()
```

## Gráfico de barras horizontales. 
Mejor para categorías con nombres largos

```python
df.groupby('Category')['Y1'].mean().plot.barh()
```

## Histográma

```python
df['Y1'].plot.hist()
```

## Gráfico de áreas

```python
#¿Cómo representa un gráfico de área el área bajo la curva de Y1 a lo largo de X?
'''
df.set_index('X', inplace=True) 
Establece la columna X como índice para que el gráfico sea en función de X.
En este caso no es necesario porque los valores de X son iguales a los del índice
'''
df['Y1'].plot.area()
plt.xlabel('X')
plt.ylabel('Y1')
```

## Boxplot
Para detectar datos atípicos

```python
#¿Qué información muestra un gráfico de caja sobre la distribución de Y1 por cada categoría, incluyendo mediana y cuartiles?

df.plot.box(column= 'Y1', by='Category')
plt.title('Gráfico de caja')
```

## Gráfico de torta
`df.plot.pie('columna')`

```python
#¿Cómo visualiza un gráfico de torta la proporción de cada categoría en el total?
df['Category'].value_counts().plot.pie(autopct='%1.1f%%')   #Muestra los porcentajes en cada rebanada del gráfico, con un decimal de precisión.
```

## Gráfico de densidad

```python
df['Y2'].plot.kde(figsize= (10,5))
```

## Gráfico hexagonal
Muestra la densidad de puntos en un espacio bidimensional.<br>
    `df.plot.hexbin(x='col1', y='col2')`

```python
# ¿Qué información proporciona un gráfico hexagonal sobre la densidad de puntos en un espacio bidimensional, y por qué es útil para grandes conjuntos de datos?
df.plot.hexbin(x='Y1', y='Y2', gridsize=30, cmap='Blues')   #Disminuir el valor de gridsize hace que los hexágonos sean más grandes, ya que hay menos hexágonos en el gráfico.
```
