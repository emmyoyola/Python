# Plotting and visualization whit Pandas

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Gráficos de linea
Sirven para series de tiempo

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2024', 
periods=1000))
ts = ts.cumsum()    #Sumas acumuladas
ts.plot()  # Grafico de línea simple con pandas
```

```python
ts.head(2)
```

## Gráficos de barra
Sirven para variables categóricas (preferiblemente)
* plot(kind='bar'): Barras verticales
* plot(kind='barh'): Barras horizontales

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=['A', 'B', 'C', 'D'])  # Gráfico de barras con pandas
df
```

```python
df.plot(kind='bar')
```

## Gráficos de Dispersión (Scatter Plots)
Sirven para identificar tendencias entre 2 variables

```python
df.plot.scatter(x='A', y='B')
```

## Gráficos de Caja (Box Plots)
Sirven para realizar comparaciones

```python
df.plot(kind='box')  # Gráfico de caja para mostrar la distribución de las columnas
plt.show()
```

## Gráficos de Histograma
Sirven para mostrar la distribución de frecuencias

```python
df['A'].plot(kind='hist', bins=8)  #Histograma para la columna A
```

```python
df.plot(kind='line')
plt.title('Gráfico de Línea')
plt.xlabel('Índice')
plt.ylabel('Valores')
plt.legend(loc='best')
plt.grid(True)
```

## Subplot
Crea múltiples gráficos en un solo lienzo

```python
df.plot(subplots=True, layout=(2, 2), figsize=(10, 8))
```

## Guardar Gráficos

```python
df.plot(kind='line')
plt.savefig('grafico_linea.png')
```
