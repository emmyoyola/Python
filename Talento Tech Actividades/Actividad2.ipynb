5. Gastos_publicidad[Gp] (Cuantitativa): Gasto en publicidad digital en USD.

#### 1. ¿Cuántos clientes pertenecen a cada categoría de Genero y Categoria_producto? Calcula las frecuencias absolutas relativas.

```python
# Frecuencia absoluta de 'Ed' (Edad)
fre_abs_ge = df['Ge'].value_counts()

# Frecuencia relativa de 'Ed'
fre_rel_ge = df['Ge'].value_counts(normalize=True)

# Frecuencia absoluta de 'Cp' (Categoría de producto)
fre_abs_cp = df['Cp'].value_counts()

# Frecuencia relativa de 'Cp'
fre_rel_cp = df['Cp'].value_counts(normalize=True)

print(f'Género Fre. Absoluta: {fre_abs_ge}')
print(f'Género Fre. Relativa: {fre_rel_ge}')
print(f'Categoría_producto Fre. Absoluta: {fre_abs_cp}')
print(f'Categoría_producto Fre. Relativa: {fre_rel_cp}')

```

#### 4. Crea un diagrama de barras para la variable Categoria_producto.

#### 5. Crea un histograma para la variable Edad. ¿Qué puedes observar sobre la distribución de edades?

```python
plt.figure(figsize=(6, 4))

# Histograma para Edad
plt.hist(df['Ed'], bins=6, color='skyblue', edgecolor='black')
plt.title('Histograma de Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
```

#### 10. Realiza un Heatmap entre todas la variables numéricas e interpreta las realaciones obtenidas.

```python
# Calcular la matriz de correlación

corr = df[["Ed", "Im", "Gp", "Fc"]].corr()
```

```python
# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de Calor de la Correlación')
plt.show()
```
