#### Datos

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sb
import scipy as scipy
```

```python
data = {'Edad':[29, 33, 51, 36, 36, 52, 40, 22, 28, 31], 'Ingreso':[84482, 67196, 68015, 62214, 48883, 95738, 69957, 20668, 74027, 50544]}
df = pd.DataFrame(data)
print("DataFrame:\n", df) 
```

#### Gráficos

```python
# Gráficos para Edad e Ingreso
plt.figure(figsize=(14, 6))

# Gráfico de líneas para Edad
'''plt.subplot(nrows, ncols, index)
nrows: Número de filas en la cuadrícula de subgráficas.
ncols: Número de columnas en la cuadrícula de subgráficas.
index: Índice de la subgráfica actual (empezando desde 1)'''
plt.subplot(1, 2, 1)
plt.plot(df.index, df['Edad'], marker='o', linestyle='-', color='blue')
plt.title('Gráfico de Líneas de Edad')
plt.xlabel('Índice')
plt.ylabel('Edad')

# Gráfico de barras para Ingreso
plt.subplot(1, 2, 2)
plt.bar(df.index, df['Ingreso'], color='green', alpha=0.7)
plt.title('Gráfico de barras de Ingreso')
plt.xlabel('Índice')
plt.ylabel('Edad')

plt.tight_layout()
plt.show()
```

```python
# Gráfico de dispersión con KDE para Edad e Ingreso (DISTRIBUCIÓN)
# La curva KDE proporciona una visualización suave de la densidad de los datos.
plt.figure(figsize=(14, 6))

# Histograma con KDE para Edad
plt.subplot(1, 2, 1)
sb.histplot(df['Edad'], kde=True, color='#F2AB6D', bins=5, stat='density')
plt.title('Histograma con KDE de Edad')
plt.xlabel('Edad')
plt.ylabel('Densidad')

# Histograma con KDE para Ingreso
plt.subplot(1, 2, 2)
sb.histplot(df['Ingreso'], kde=True, color='#8ABF7C', bins=5, stat='density')
plt.title('Histograma con KDE de Ingreso')
plt.xlabel('Ingreso')
plt.ylabel('Densidad')

# Mostrar gráficos
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
# Gráfico de dispersión (Relaciones entre variables)
plt.scatter(df['Edad'], df['Ingreso'], color='blue')
plt.title('Gráfico de Dispersión: Edad vs Ingreso')
plt.xlabel('Edad')
plt.ylabel('Ingreso')

plt.grid(True)
plt.show()
```

```python
# Crear el gráfico de dispersión con línea de tendencia
plt.figure(figsize=(6, 4))
#Ajustar y trazar una línea de regresión lineal automáticamente.
sb.regplot(x='Edad', y='Ingreso', data=df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'}) 
plt.title('Gráfico de Dispersión con Línea de Tendencia: Edad vs Ingreso')
plt.xlabel('Edad')
plt.ylabel('Ingreso')
plt.grid(True)
plt.show()
```

##### **Pregunta 13:** Calcula los valores atípicos para la variable ingreso usando el rango intercuartílico (IQR). ¿Cuántos valores atípicos detectas?

```python
iqr_e= df['Edad'].quantile(0.75) - df['Edad'].quantile(0.25)
iqr_i = df['Ingreso'].quantile(0.75) - df['Ingreso'].quantile(0.25)

print(f'El rango intercuartílico (IQR) para Edad es {iqr_e:.2f}')
print(f'El rango intercuartílico (IQR) para Ingreso es {iqr_i:.2f}')
```

```python
lim_inf_ingresos = (df['Ingreso'].quantile(0.25)-1.5)*iqr_i
lim_sup_ingresos = (df['Ingreso'].quantile(0.75)+1.5)*iqr_i
```

```python
lim_inf_ingresos
```

```python
df[(df['Ingreso']<lim_inf_ingresos) | (df['Ingreso']>lim_sup_ingresos)].shape[0]
```

Rta/: Usando el IQR se detectaron 10 valores atípicos
