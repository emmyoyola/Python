# **Análisis Gráfico**
- Los gráficos que cubriremos son esenciales para análisis o de datos: **histogramas, diagramas de caja (boxplots), gráficos de barras, gráficos de dispersión, gráficos de líneas y gráficos de violín.**

- Usaremos las bibliotecas más comunes para visualización en Python, como **matplotlib** y **seaborn**.

1.**Instalar y cargar las bibliotecas necesarias**

```python
# importar las bibliotecas
import numpy as np
import pandas as pd
import seaborn as sns                      
import matplotlib.pyplot as plt 
import plotly.express as px

from sklearn.datasets import load_iris  # Contiene df de muestra
```

2. **Cargar las datos**

```python
# Datos de ejemplo
Edad = [2, 4, 6, 8, 10, 4]
Genero = ['M', 'F', 'M', 'F', 'M', 'F']
Peso = [45, 35, 86, 57, 59, 43]

df = pd.DataFrame({'Edad': Edad, 'Genero': Genero, 'Peso': Peso})
df.head()
```

## Histogramas

- Un histograma muestra la distribución de una variable continua dividiéndola en intervalos.



```python
# Con Matplotlib
plt.figure(figsize=(8, 6)) #Configuración del tamaño de la gráfica
plt.hist(Edad, bins=5,  color='skyblue', edgecolor='black')
plt.title('Histograma de Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Con seaborn
sns.histplot(data=df, x='Edad')
plt.title('Histograma de Edad')
plt.show()
```

```python
# Con pandas
df['Edad'].plot.hist()
plt.title('Histograma de Edad')
```

```python
# Con plotly
fig = px.histogram(data_frame=df, x='Edad', nbins=5, title='Histograma de Edad')
fig.show()
```

## Diagrama de Caja (Boxplot)

Un boxplot muestra la distribución de los datos resaltando los cuartiles y valores atípicos.

***Importante***
- Datos atípicos
- mediana
- sesgo



```python
# Con Matplotlib
plt.figure(figsize=(14, 2))
plt.boxplot(x=Peso, vert=False)
plt.title('Diagrama de Caja de Peso')
plt.show()
```

```python
# Con seaborn
plt.figure(figsize=(14, 2))
sns.boxplot(x=Peso, color='tomato')
plt.title('Diagrama de Caja de Peso')
plt.xlabel('Peso')
plt.show()
```

```python
# Con pandas
df['Peso'].plot.box()
plt.title('Diagrama de Caja de Peso')
```

```python
# Con plotly
plt.figure(figsize=(14, 2))
px.box(x=df['Peso'], title='Diagrama de Caja de Peso')
```

## Gráfico de Densidad

- Un gráfico de densidad es una visualización de la estimación de la función de densidad.


```python
# Con seaborn
plt.figure(figsize=(8, 6))
sns.kdeplot(df['Edad'], fill=True, color='blue')
plt.title('Gráfico de Densidad de Edad')
plt.xlabel('Edad')
plt.ylabel('Densidad')
plt.show()
```

```python
# Con pandas
df['Edad'].plot.kde()
plt.title('Gráfico de Densidad de Edad')
plt.show()
```

##  Gráfico de Barras

- Un gráfico de barras es útil para visualizar la frecuencia de datos categóricos.


```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Edad', data = df, hue='Edad', palette='Set2')
plt.title('Gráfico de Barras de Categorías')
plt.xlabel('Categoría')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Sacar tabla de frecuencia
a = df['Genero'].value_counts()
rec_genero = pd.DataFrame(a)
```

```python
# Con Matplotlib
plt.figure(figsize=(6,4))
plt.bar(x=rec_genero.index, height=rec_genero['count'])
plt.title('Recuento de genero')
```

```python
# Con seaborn
plt.figure(figsize=(6,4))
sns.barplot(data= rec_genero, x=rec_genero.index, y='count')
plt.title('Recuento de genero')
```

```python
# Con pandas
rec_genero['count'].plot(kind='bar')
plt.title('Recuento de genero')
plt.show()
```

```python
rec_genero['count'].plot.bar()
plt.title('Recuento de genero')
plt.show()
```

```python
rec_genero['count'].plot.barh()
plt.title('Recuento de genero')
plt.show()
```

```python
# Con plotly
fig = px.bar(rec_genero, x=rec_genero.index, y='count', title= 'Recuento de genero')
fig.show()
```

```python
# Con plotly
fig = px.bar(df, x='Genero', y='Peso', title='Satisfacción del Entorno Laboral por Nivel Educativo')
fig.show()
```

## Gráfico de Dispersión (Scatter Plot)

- Un gráfico de dispersión muestra la relación entre dos variables continuas.


```python
# Con matplotlib
plt.figure(figsize=(8, 6))
plt.scatter(df['Edad'], df['Peso'], color='purple')
plt.title('Gráfico de Dispersión: Edad vs Peso')
plt.xlabel('Edad')
plt.ylabel('Peso')
plt.show()
```

```python
# Con seaborn
sns.scatterplot(data=df, x='Edad', y='Peso')
plt.title('Gráfico de Dispersión: Edad vs Peso')
plt.show()
```

```python
# Con pandas
df.plot.scatter(x='Edad', y='Peso')
plt.title('Gráfico de Dispersión: Edad vs Peso')
plt.show()
```

```python
px.scatter(data_frame=df, x='Edad', y='Peso', title='Gráfico de Dispersión: Edad vs Peso')
```

## Gráfico de Violín (Violin Plot)

- El gráfico de violín es una combinación de boxplot y estimación de la densidad de probabilidad.


```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Genero', y='Edad', data=df, hue='Genero',palette='muted')
plt.title('Gráfico de Violín de Edad por Genero')
plt.xlabel('Genero')
plt.ylabel('Edad')
plt.show()
```

```python
# Con plotly
fig = px.violin(data_frame=df, x='Genero', y='Edad', color='Genero', box=True, points='all')
fig.show()
```

## Gráfico de Líneas

- Un gráfico de líneas es útil para mostrar tendencias a lo largo del tiempo o secuencias ordenadas.


```python
# Con matplotlib
plt.figure(figsize=(8, 6))
plt.plot(df['Edad'], color='darkred')
plt.title('Gráfico de Líneas de Edad')
plt.xlabel('Índice')
plt.ylabel('Edad')
plt.show()
```

```python
# Con seaborn
sns.lineplot(data=df,x=df.index, y='Edad')
plt.title('Gráfico de Líneas de Edad')
plt.xlabel('Índice')
plt.show()
```

```python
# Con pandas
df['Edad'].plot.line()
plt.title('Gráfico de Líneas de Edad')
plt.xlabel('Índice')
plt.ylabel('Edad')
plt.show()
```

```python
# Con plotly
px.line(df['Edad'], title='Gráfico de Líneas de Edad')
```

## Gráfico de Correlación (Heatmap)

- El heatmap es útil para visualizar la relación entre múltiples variables numéricas.



```python
# Load the iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Calcular la matriz de correlación
corr = df.corr()
print(corr)

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de Calor de la Correlación')
plt.show()
```
